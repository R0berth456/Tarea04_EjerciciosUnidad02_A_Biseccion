Primer Literal

```python
def sign(x: float) -> int:
    if x > 0:
        return 1
    elif x < 0:
        return -1
    else:
        return 0
```

```python
from typing import Callable
def bisection(
    a: float, b: float, *, equation: Callable[[float], float], tol: float, N: int
) -> tuple[float, int] | None:
    i = 1

    assert a < b, "a not lower than b, the interval is not valid."

    assert (
        equation(a) * equation(b) < 0
    ), "The function does not change sign over the interval."

    Fa = equation(a)
    p = a
    for i in range(N):
        p = a + (b - a) / 2
        FP = equation(p)
        if FP == 0 or (b - a) / 2 < tol:
            return p, i

        if sign(Fa) * sign(FP) > 0:
            a = p
            Fa = FP
        else:
            b = p

    return p, i
```

```python
r, i = bisection(a = 0, b = 1, equation = lambda x : x**3 - 7*x**2 + 14*x - 6, tol = 10**(-2), N = 20 )
print(f'raiz = {r} en i = {i}')
```

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return (x + 3) * (x + 1)**2 * x * (x - 1)**3 * (x - 3)

x = np.linspace(-4, 4, 400)
y = f(x)

plt.figure(figsize=(10, 6))
plt.plot(x, y, label=r'$f(x) = (x+3) \cdot (x+1)^2 \cdot x \cdot (x-1)^3 \cdot (x-3)$')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Gr치fica de la funci칩n $f(x) = (x+3) \cdot (x+1)^2 \cdot x \cdot (x-1)^3 \cdot (x-3)$')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.legend()
plt.show()
```
